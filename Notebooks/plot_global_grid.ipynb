```python
import matplotlib.pyplot as plt
import matplotlib.collections
# from matplotlib.collections import LineCollection
import numpy as np


def showMeshPlot(nodes, elements, values):

    y = nodes[:,0]
    z = nodes[:,1]

    def quatplot(y,z, quatrangles, values, ax=None, **kwargs):

        if not ax: ax=plt.gca()
        yz = np.c_[y,z]
        verts= yz[quatrangles]
        pc = matplotlib.collections.PolyCollection(verts, **kwargs)
        pc.set_array(values)
        ax.add_collection(pc)
        ax.autoscale()
        return pc

    fig, ax = plt.subplots()
    ax.set_aspect('equal')

    pc = quatplot(y,z, np.asarray(elements), values, ax=ax,
             edgecolor="crimson")
    fig.colorbar(pc, ax=ax)
    ax.plot(y,z, marker="o", ls="", color="crimson")

    ax.set(title='This is the plot for: quad', xlabel='Y Axis', ylabel='Z Axis')

    plt.show()

nodes = np.array([[0,0], [0,0.5],[0,1],[0.5,0], [0.5,0.5], [0.5,1], [1,0],
                  [1,0.5],[1,1]])
elements = np.array([[0,3,4,1],[1,4,5,2],[3,6,7,4],[4,7,8,5]])
stresses = np.array([1,2,3,4])

showMeshPlot(nodes, elements, stresses)
```

# Plotting an unstructured mesh
We want to plot the mesh for the spectral element dynamical core.

## From the grid description file
These files (e.g., `ne30np4_091226_pentagons.nc`) have variables that explicitly define the grid cell corners. Using these variables, we can construct all the necessary polygons that make up the mesh and plot them as a collection.

```python
se_ds = xr.open_dataset("/Volumes/Samsung_T5/ne30np4_091226_pentagons.nc")

# grid corner variables
gclon = se_ds['grid_corner_lon'].values
gclat = se_ds['grid_corner_lat'].values

```

```python
verts = []
for i in range(len(gclon)):
    v = np.stack([gclon[i, :], gclat[i, :]], 1)
    if i == 0:
        print(f"Shape of v: {v.shape}")
    verts.append(v)
# verts is a list of (nv, 2) arrays, where nv is number of vertieces

fig, ax = plt.subplots(figsize=(10,5), subplot_kw={'projection':ccrs.PlateCarree()})
# p = PatchCollection(polygon, color='none', alpha=0.4, ec='k')
p = PolyCollection(verts, edgecolor="gray", facecolor='none', linewidths=1)
p.set_transform(ccrs.PlateCarree())
ax.add_collection(p)
# ax.set_global()
ax.set_extent([-120, -90, 15, 50], crs=ccrs.PlateCarree())
ax.coastlines(color='r')

# I think we see the "stitching" across faces here.
```
