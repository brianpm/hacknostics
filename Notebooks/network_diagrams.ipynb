```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

def create_network_diagram(nodes, connections, keywords=None):
    # Create a new graph
    # G = nx.Graph()
    # G = nx.MultiDiGraph()

    n = len(nodes)
    m = len(connections)
    seed = 555
    G = nx.Graph()

    # Add nodes
    G.add_nodes_from(nodes)

    # Add edges with keywords
    for connection in connections:
        if len(connection) == 3:
            person1, person2, keyword = connection
        else:
            person1, person2 = connection
            keyword=None
        G.add_edge(person1, person2, keyword=keyword)

    # # Create a color map for keywords
    # color_map = {keyword: plt.cm.get_cmap('tab10')(i/len(keywords)) 
    #              for i, keyword in enumerate(keywords)}

    # Find the minimum spanning tree
    T = nx.minimum_spanning_tree(G)

    # Draw the network
    pos = nx.spring_layout(G, 4/np.sqrt(n))  # Position nodes using Fruchterman-Reingold force-directed algorithm

    # pos = nx.circular_layout(G)

    # Generate layout for visualization
    # pos = nx.kamada_kawai_layout(G)

    plt.figure(figsize=(12, 8))

    # Draw nodes
    # nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')
    # nx.draw_networkx_labels(G, pos)

    # Draw edges
    # for (node1, node2, data) in G.edges(data=True):
    #     nx.draw_networkx_edges(G, pos, edgelist=[(node1, node2)], width=2)
    #                         #    edge_color=[color_map[data['keyword']]], width=2)

    # nx.draw_networkx_edges(T, pos, edge_color="green", width=2)

    # Create a legend
    # legend_elements = [plt.Line2D([0], [0], color=color_map[kw], lw=4, label=kw) 
    #                    for kw in keywords]
    # plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1))


    nx.draw_networkx(G, pos=pos)

    plt.title("Network Diagram")
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage
nodes = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank']
keywords = ['Work', 'Family', 'Friend', 'School']
connections = [
    ('Alice', 'Bob', 'Work'),
    ('Alice', 'Charlie', 'Friend'),
    ('Bob', 'David', 'School'),
    ('Charlie', 'Eve', 'Family'),
    ('David', 'Eve', 'Work'),
    ('Frank', 'Alice', 'School'),
    ('Frank', 'Eve', 'Friend')
]

create_network_diagram(nodes, connections, keywords)
```
